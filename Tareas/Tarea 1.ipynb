```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
from tqdm import tqdm
import random as rnd
```

```python
class Particle:
    
    def __init__(self, r0, v0, a0, t, m=1, radius=2., Id=0):
        
        self.dt = t[1] - t[0]
        
        
        # Atributos instantaneos
        self.r = r0
        self.v = v0
        self.a = a0
        
        self.m = m
        self.radius = radius
        self.Id = Id
        
        
        self.p = self.m*self.v
        
        self.f = self.m*self.a
        
        # Historial
        
        self.R = np.zeros((len(t),len(r0)))
        self.V = np.zeros_like(self.R)
        self.A = np.zeros_like(self.R)
        
        self.F = np.zeros_like(self.R)
        
        self.P = np.zeros_like(self.R)
    
        # Fisica
        self.K = 100.
        
        self.VEk = np.zeros(len(t))
        
    def Evolution(self,i):
        
        
        self.SetPosition(i)
        self.SetVelocity(i)
        
        self.a = self.f/self.m
        
        self.v += self.dt*self.a
        self.r += self.dt*self.v
        
        
    def CalculateForce(self,p):
        
        d = np.linalg.norm(self.r - p.r)
        
        compresion = self.radius + p.radius - d
        
        if compresion >= 0:
            
            Fn = self.K * compresion**3
            
            self.n = (self.r - p.r)/d
            
            self.f = Fn*self.n
            
           # print(self.f)
            
        
    # Setter
    def SetPosition(self,i):
        self.R[i] = self.r
    
    def SetVelocity(self,i):
        self.VEk[i] = 0.5*self.m*np.dot(self.v,self.v)
        self.V[i] = self.v
        self.P[i] = self.m*self.v
        
    
    # Getter
    def GetPosition(self,scale=1):
        return self.R[::scale]
    
    def GetVelocity(self,scale=1):
        return self.V[::scale]
 
    def GetMomentum(self,scale=1):
        return self.P[::scale]
```

```python
def GetParticles(N,t):
    
    Particles = []

    for i in range(N):
        r0 = np.array([float(rnd.randrange(-10,10)),float(rnd.randrange(-10,10))])
        v0 = np.array([float(rnd.randrange(-5,5)),float(rnd.randrange(-5,5))])
        a0 = np.array([0.,0.])
        Particles.append(Particle(r0,v0,a0,t))
    
    return Particles
```

```python
dt = 0.001
tmax = 10
t = np.arange(0,tmax,dt)
Particles = GetParticles(4,t)
```

```python
def RunSimulation(t,Particles):
    
    for it in range(len(t)):
        
        for i in range(len(Particles)):
            for j in range(len(Particles)):
                if i!=j:
                    Particles[i].CalculateForce(Particles[j])
        
        
        
        for i in range(len(Particles)):
            Particles[i].Evolution(it)
            
            
    return Particles
```

```python
Particles = RunSimulation(t,Particles)
```

## Punto 7.

Considere la ecuación diferencial $$ \frac{\text{d}u}{\text{d}t} = \alpha u ,\, u(0)=u_0.$$ 
- Muestre que aplicando iterativamente se obtiene que $$u_k = (1+\alpha \Delta t)^k u_0$$ para el caso en que $\alpha < 0$.


**Solución:**

Teniendo en cuenta el método de Euler, se sabe que $$u_{i+1} = u_{i} + \Delta t (\alpha u_{i})$$ que se puede reescribir como $$u_{k} = (1 + \alpha \Delta t)u_{k-1}.$$ Sin embargo, se tiene, por la segunda igualdad que $$u_{k} = (1 + \alpha\Delta t)^2u_{k-2}$$ usando estos pasos de manera iterada $k-1$ veces, se obtiene que $$u_k = (1 + \alpha\Delta t)^ku_0.$$

*** 
- Muestre que la solución numérica oscilará si $\Delta t > -\frac{1}{\alpha}$. 


**Solución:**

Usando el hecho de que $\alpha < 0$. Se puede llegar a que $$ \alpha \Delta t < -1 $$ sumando uno a ambos lados $$ 1 + \alpha t < 0 $$ Esto significa que la expresión $(1 + \alpha \Delta t)^k$ será positiva o negativa dependiendo de si $k$ es par o impar, respectivamente.

***

- Ajuste $\alpha = -1$ y muestre las soluciones oscilatorias para $\Delta t = 1.1, 1.5 \text{ y } 1.9 \text{ s}$. Por otro lado, la solución exacta nunca oscila $u(t) = e^{\alpha t}$.

**Solución:**

```python
import numpy as np
import matplotlib.pyplot as plt

dt = [1.1,1.5,1.9]
N = 100
alpha = -1
u = [np.zeros(N),np.zeros(N),np.zeros(N)]

ax = plt.subplot(1,1,1)



for j in range(len(dt)):
    for i in range(N-1):
        if i == 0:
            u[j][i] = 1
        u[j][i+1] = (1 + alpha*dt[j])*u[j][i]
    t = np.arange(0,10,dt[j])
    ax.plot(t,u[j][:len(t)],label=r"Solución con $\Delta t =$ %s s"%(dt[j]),linestyle="dashed")
tlins = np.linspace(0,10)
ax.plot(tlins,np.exp(alpha*tlins),label=r"Solución $u(t) = e^{\alpha t}$")

plt.legend()
```
