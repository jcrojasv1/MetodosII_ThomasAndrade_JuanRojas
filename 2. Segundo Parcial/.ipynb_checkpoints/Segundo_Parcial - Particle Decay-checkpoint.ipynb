```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

$\verb|Particle decay:|$ Una partícula fundamental dada $A$ puede desintegrarse a una partícula $B$ con una tasa de decaimiento ($\lambda_{A} = 0.5$ particulas$/$h). A su vez, la partícula $B$ puede desintegrarse a una partícula $C$ con una tasa de decaimiento ($\lambda_{B} = 0.3$ particulas/h). La partícula $C$ es estable y la desintegración directa de $A \to C$ está prohibida. Las tasas
de decaimiento para una sola especie tiene una forma conocida $\frac{dN}{dt} = −\lambda N$, sin embargo, para decaimientos sucesivos $(A \to B \to C)$ aparece un sistema acoplado de ecuaciones diferenciales.

```python
lambda_A = 0.5  # Partículas/Hora
lambda_B = 0.3  # Partículas/Hora


def EqSystem(r, t):
    Na, Nb, Nc = r
    
    dNadt = - lambda_A * Na
    dNbdt = - lambda_B * Nb + lambda_A * Na
    dNcdt = lambda_B * Nb
    
    return np.array([dNadt, dNbdt, dNcdt])
```

```python
def Increment(f,r,t,h):
    state1 = np.array([r[0],r[1], r[2]])
    K1 = f(state1,t)
    
    state2 = np.array([r[0]+0.5*h*K1[0],r[1]+0.5*h*K1[1], r[2]+0.5*h*K1[2]])
    K2 = f(state2,t+0.5*h)
    
    state3 = np.array([r[0]+0.5*h*K2[0],r[1]+0.5*h*K2[1], r[2]+0.5*h*K2[2]])
    K3 = f(state3,t+0.5*h)
    
    state4 = np.array([r[0]+h*K3[0],r[1]+h*K3[1], r[2]+h*K3[2]])
    K4 = f(state4, t+h)
    
    Deltaf = h*(K1+2*K2+2*K3+K4)/6
    
    return r+Deltaf
```

```python
def Adaptativo(f,x0,t, e = 1e-4):
    
    h = t[1] - t[0]
    
    tf = t[-1]
    t = t[0]
    
    r = x0
    
    Time = np.array([t])
    Vector = r
    
    Steps = np.array([h])
    
    while t < tf:
        
        r1 = Increment(f,r,t,h)
        r2 = Increment(f,r1,t + h,h)
        rt = Increment(f,r,t,2*h)
    
        error = np.linalg.norm(r2 - rt)
        
        if error < e:
            t = t + 2*h
            r = rt
            Time = np.append(Time,t)
            Steps = np.append(Steps,h)
            Vector = np.vstack((Vector,r))
            
        delta = 0.85 * (e/error)**(1/4)
        h = h*delta

    return Vector, Time, Steps
```

```python
t = np.linspace(0, 10, 50)
r = np.array([500, 0, 0])
solution = odeint(EqSystem, r, t)
```

A continuación se muestra la situación de decaimiento integrada con el método de Integrate Odeint de Scipy. 

```python
plt.plot(t,solution[:,0],marker='.', color = "orange", label = "Particle A")
plt.plot(t,solution[:,1],marker='.', color = "grey", label = "Particle B")
plt.plot(t,solution[:,2],marker='.', label = "Particle C")
plt.title("Situación Integrada con Scipy Integrate Odeint")
plt.grid()
plt.legend()
```

Ahora se aprecia la comparativa visual entre Integrate Odeint de Scipy y el resultado obtenido con el integrador adaptativo RK4. Las gráficas de Odeint están desplazadas dos unidades hacia abajo, para que la comparación visual sea más evidente.
